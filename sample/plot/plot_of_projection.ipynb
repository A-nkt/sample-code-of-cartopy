```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
```

## 正距円筒図法

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree(central_longitude=180))
ax.coastlines()
```
